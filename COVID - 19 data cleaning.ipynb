```python
# Drop completely empty unnamed columns
dataset.dropna(axis=1, how='all', inplace=True)

# Rename unnamed columns if they contain useful data
dataset.columns = [col if not col.startswith("Unnamed") else f"Column_{i}" 
                   for i, col in enumerate(dataset.columns)]
```

```python

# Fill categorical columns with mode
for col in dataset.select_dtypes(include='object').columns:
    mode = dataset[col].mode()
    if not mode.empty:
        dataset[col] = dataset[col].fillna(mode[0])

# Fill numerical columns with median
for col in dataset.select_dtypes(include='number').columns:
    median = dataset[col].median()
    dataset[col] = dataset[col].fillna(median)

# Optional: Fill datetime columns with the most frequent date
for col in dataset.select_dtypes(include='datetime').columns:
    mode = dataset[col].mode()
    if not mode.empty:
        dataset[col] = dataset[col].fillna(mode[0])
```
