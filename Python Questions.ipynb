```python
# Write a function to check whether a number is prime
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2,n):
        if n % i == 0:
            return False
    return True

print(is_prime(7))

print(is_prime(10))
```

```python
# Write a function to find factorial of a number.
def factorial_iterative(n: int):
    if n < 0:
        raise ValueError("factorial is not defined for negative numbers")
    result=1
    for i in range(1, n+1):
        result *= i
    return result

print(factorial_iterative(5))
```

```python
# Write a function to return the maximum of three numbers.
def max_of_three(a: int, b: int, c: int):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

print(max_of_three(20,25,30))
```

```python
# Write a function to count vowels in a string

def count_vowels(s: str) -> int:
    vowels = "aeiou"
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count

print(count_vowels("parshant mishra"))
```

```python
# Count frequency of elements in a list.
def second_largest(nums):
    nums = list(set(nums))  # remove duplicates
    nums.sort()
    return nums[-2] if len(nums) >= 2 else None

print(second_largest([10, 20, 4, 45, 99]))  # Output: 45
```

```python
# Find the sum of first n natural numbers
def sum_natural_numbers_loop(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total

n = 10
print("Sum of first", n, "natural numbers is:", sum_natural_numbers_loop(n))
```

```python
def fibonacci_series(n):
    fib = []
    a, b = 0, 1
    for _ in range(n):
        fib.append(a)
        a, b = b, a + b
    return fib

# Example usage:
n = 10
print("Fibonacci series up to", n, "terms:", fibonacci_series(n))
```
