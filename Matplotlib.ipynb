```python
import matplotlib.pyplot as plt
import numpy as np

x=["python", "c", "c++", "java"]
y=[90,80,60,85]
z=[20,30,40,35]
c=[40, 50, 60, 70] 

width=0.2
p=np.arange (len(x))
p1=[j+width for j in p]
p2 = [j + width * 2 for j in p] 

plt.title("Netmax", fontsize=15)
plt.xlabel("languages", fontsize=15)
plt.ylabel("No", fontsize=15)

plt.bar(p,y, width, color="r", label="popularity")
plt.bar(p1,z, width, color="y", label="netmax")
plt.bar(p2, c, width, color="g", label="tech")

plt.xticks(p+width/2,x)    # Replace numeric position with subject name.
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=["python", "c", "c++", "java"]
y=[90,80,60,85]
z=[20,30,40,35]

width=0.2
p=np.arange (len(x))
p1=[j+width for j in p]

plt.title("Netmax", fontsize=15)
plt.xlabel("languages", fontsize=15)
plt.ylabel("No", fontsize=15)

plt.barh(p,y, width, color="r", label="popularity")  # HORIZONTAL - use h.
plt.barh(p1,z, width, color="y", label="popularity")

plt.xticks(p+width/2,x)
plt.legend()
plt.show()
```

# Stack Plot

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
area=[2,3,5,4,6]
area1=[5,4,8,2,3]
area2=[8,4,5,6,2]

l=["area","area1","area2"]

plt.stackplot(x,area,area1,area2,labels=l)

plt.title("netmax")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.legend(loc=2)
plt.show()
```

```python
import pandas as pd
import numpy as np

data=[10,20,30,40,50,60,150]
print(data)

data = np.array([10,20,30,40,50,60,150])
print(data)
```

```python
Q1=np.percentile(data, 25)
Q1
```

```python
Q3=np.percentile(data, 75)
Q3
```

```python
IQR = Q3- Q1
IQR
```

```python
 lower_bound = Q1- 1.5 * IQR
lower_bound
```

```python
upper_bound = Q3 + 1.5 * IQR
upper_bound
```

```python
outlier_mask = (data < lower_bound) | (data > upper_bound)
outlier_mask
```

```python
outliers = data[outlier_mask]
outliers
```

```python
data_no_outliers = data[~outlier_mask]
data_no_outliers
```
