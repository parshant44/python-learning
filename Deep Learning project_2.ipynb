# House Price

```python
import pandas as pd
dataset=pd.read_csv(r"C:\Users\ADMIN\Documents\House_price_dataset.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.drop(columns = ['date','waterfront','view','condition','statezip','yr_built','yr_renovated','street'],inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()

dataset['city'] = encoder.fit_transform(dataset['city'])


dataset['country'] = encoder.fit_transform(dataset['country'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=dataset[["bedrooms","bathrooms","sqft_living","sqft_lot","floors","sqft_above","sqft_basement","city","country"]]
y=dataset["price"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train_trf = scaler.fit_transform(x_train)
x_test_trf = scaler.transform(x_test)
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential()

model.add(Dense(64,activation='sigmoid',input_dim=11))
model.add(Dense(32,activation='relu'))
Dropout(0.2)
model.add(Dense(32,activation='relu'))
Dropout(0.2)
model.add(Dense(32,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# Define early stopping callback
early_stopping = EarlyStopping(
    monitor='val_loss',  # Metric to monitor
    patience=5,          # Number of epochs with no improvement before stopping
    restore_best_weights=True  # Restore model weights from the best epoch
)
```

```python
history = model.fit(x_train_trf,y_train,batch_size=20,epochs=50,verbose=1,validation_split=0.2,callbacks=[early_stopping])
```
