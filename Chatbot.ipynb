```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder
```

```python
data = {
    "question": [
        "I have fever",
        "I have headache",
        "I am feeling cold",
        "I have stomach pain",
        "I am coughing",
        "I feel tired",
        "I have sore throat",
        "I have body pain",
        "I feel dizzy",
        "I have nausea"
    ],
    "answer": [
        "You may have a fever. Please take rest and consult a doctor if it continues.",
        "Headache can be caused by stress or dehydration. Drink water and rest.",
        "Cold may indicate viral infection. Keep yourself warm.",
        "Stomach pain can be due to indigestion. Avoid spicy food.",
        "Cough may be due to cold or allergy. Drink warm fluids.",
        "Fatigue can be caused by stress or lack of sleep.",
        "Sore throat can be relieved by warm salt water gargle.",
        "Body pain may be due to weakness or fever.",
        "Dizziness can be caused by low blood pressure.",
        "Nausea may be due to gastric issues. Eat light food."
    ]
}

df = pd.DataFrame(data)
df
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(df["question"])

X = tokenizer.texts_to_sequences(df["question"])

X = pad_sequences(X, padding='post')

max_len = max(len(seq) for seq in X)
```

```python
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(df["answer"])

y = tf.keras.utils.to_categorical(y)
```

```python
model = Sequential()
model.add(Embedding(input_dim=1000, output_dim=64, input_length=X.shape[1]))
model.add(LSTM(64, return_sequences=False))
model.add(Dropout(0.3))
model.add(Dense(32, activation='relu'))
model.add(Dense(y.shape[1], activation='softmax'))

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
model.build(input_shape=(None,max_len))
model.summary()
```

```python
model.fit(
    X, y,
    epochs=100,
    batch_size=2,
    verbose=1
)
```

```python
def chatbot_response(user_input):
    seq = tokenizer.texts_to_sequences([user_input])
    padded = pad_sequences(seq, maxlen=X.shape[1], padding='post')
    
    prediction = model.predict(padded)
    response_index = np.argmax(prediction)
    
    return label_encoder.inverse_transform([response_index])[0]
```

```python
while True:
    user_input = input("You: ")
    if user_input.lower() in ["exit", "quit"]:
        print("Chatbot: Take care! ğŸ‘‹")
        break
    
    response = chatbot_response(user_input)
    print("Chatbot:", response)
```

```python
model.save("Chatbot_Model.h5")
```
