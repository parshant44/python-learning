```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score,root_mean_squared_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import StandardScaler
```

```python
dataset=pd.read_csv(r"C:\Users\ADMIN\Documents\cardekho.csv")
dataset.head()
```

```python
dataset.duplicated()
```

```python
dataset.isnull().sum()
```

```python
dataset["mileage(km/ltr/kg)"]=dataset["mileage(km/ltr/kg)"].fillna(dataset["mileage(km/ltr/kg)"].mean())
```

```python
dataset["engine"]=dataset["engine"].fillna(dataset["engine"].mean())
```

```python
dataset['max_power'] = pd.to_numeric(dataset['max_power'], errors='coerce')
```

```python
dataset["max_power"]=dataset["max_power"].fillna(dataset["max_power"].mean())
```

```python
dataset.fillna({"seats": dataset["seats"].mode()[0]}, inplace=True)
dataset
```

```python
dataset.describe()
```

```python
#x = dataset[["year","km_driven","mileage(km/ltr/kg)","engine"]]
#y = dataset["selling_price"] 


x=dataset.drop(columns=["name","selling_price"])
y=dataset["selling_price"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

x_train=pd.get_dummies(x_train,columns=["fuel","transmission","owner","seller_type"],dtype=int)
x_test=pd.get_dummies(x_test,columns=["fuel","transmission","owner","seller_type"],dtype=int)
x_test = x_test.reindex(columns=x_train.columns, fill_value=0)

x_train["max_power"] = pd.to_numeric(x_train["max_power"], errors='coerce')

```

```python
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
model=LinearRegression()
model.fit(x_train_scaled,y_train)
y_pred=model.predict(x_test_scaled)

```

```python
mse=mean_squared_error(y_test,y_pred)
print(f"MSE: {mse}")

print("  ")

r2=r2_score(y_test,y_pred)
print(f"R2_Score  : {r2}")
```

```python
#random Forest model
model1=RandomForestRegressor(n_estimators=100,max_depth=5,random_state=42)
model1.fit(x_train_scaled,y_train)
y_pred1=model1.predict(x_test_scaled)

mse1=mean_squared_error(y_test,y_pred1)
print(f"MSE: {mse1}")

print("  ")

r2r=r2_score(y_test,y_pred1)
print(f"R2_Score  : {r2r}")
```

```python
#Decision tree model
model2=DecisionTreeRegressor(max_depth=10,random_state=42)
model2.fit(x_train_scaled,y_train)
y_pred2=model2.predict(x_test_scaled)

mse2=mean_squared_error(y_test,y_pred2)
print(f"MSE: {mse2}")

print("  ")

r2d=r2_score(y_test,y_pred2)
print(f"R2_Score  : {r2d}")
```
