# break

```python
i=1
while i<=5:
    print(i)
    if(i==3):
        break
    i+=1
    
```

# continue

```python
i=0
while i<=10:
    if(i==7):
        i+=1
        continue
    print(i)
    i+=1
```

```python
# to print odd no.
i=0
while i<=10:
    if(i%2 ==0):
        i+=1
        continue 
    print(i)
    i+=1
```

```python
# to print even no.
i=0
while i<=10:
    if(i%2 != 0):
        i+=1
        continue
    print(i)
    i+=1    
    
```

# For loop - are used to sequential traversal. for traversing list, string, tuples etc.

```python
c=[1,2,3,4]

for val in c:
    print(val)
```

```python
p="apk"

for ck in p:
    print(ck)
else:
    print("thank you")
```

```python
p="apk"

for ck in p:
    if(ck=="p"):
        print("p found")
        break
    print(ck)
else:
    print("thank you")
```

```python
# SEARCH X IN LIST

c=[1,5,1,4,2,4,2,6]
x=6

l=0
for kl in c:
    if(kl==x):
        print("number found at l",l)
        l+=1
```
