# Mean Absolute Deviation - gives idea about variability.

```python
ab = np.array([25,60,45,20,14])
cd = np.array([18,45,14,54,65]) 
no = np.array([1,2,3,4,5])
```

```python
mean = np.mean(cd)
mean
```

```python
ab = np.sum(np.abs(ab-mean))/len(ab)
cd = np.sum(np.abs(cd-mean))/len(cd)
```

```python
ab
cd
```

```python
ab = [65, 70, 75, 80, 85]
cd = [60, 68, 72, 78, 90]
no = [1, 2, 3, 4, 5]  
```

```python
plt.figure(figsize=(7,2))
plt.scatter(ab, no, label="ab")
plt.scatter(cd, no, color="red", label="cd")
plt.plot([70]*len(no), no, c="yellow", label="mean")
plt.legend()
plt.show()
```
