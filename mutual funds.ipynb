```python
import requests
from prettytable import PrettyTable

url = "https://indian-stock-exchange-api2.p.rapidapi.com/mutual_funds"
headers = {
    "x-rapidapi-key": "your_actual_key_here",
    "x-rapidapi-host": "indian-stock-exchange-api2.p.rapidapi.com"
}

response = requests.get(url, headers=headers)
record = response.json().get("Debt", {}).get("Floating Rate", [])

table = PrettyTable()
table.field_names = ["S.No", "Fund Name", "Asset Size", "1M Return", "3M Return", "6M Return", "Star Rating"]

for i in range(len(record)):
    fund = record[i]
    table.add_row([
        i + 1,
        fund.get("fund_name", ""),
        fund.get("asset_size", ""),
        fund.get("1_month_return", ""),
        fund.get("3_month_return", ""),
        fund.get("6_month_return", ""),
        fund.get("star_rating", "")
    ])

print(table)
```
