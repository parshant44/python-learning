```python
"""
Project: Resume Tracking System (ATS)
Rule: Resume will be shortlisted only if skill match >= 80%
Author: Your Name
"""
```

```python
job_title = "Data Analyst"

job_required_skills = [
    "python",
    "sql",
    "excel",
    "machine learning",
    "statistics"
]

print("Job Title:", job_title)
print("Required Skills:", job_required_skills)
```

```python
resume_text = """
Name: Rahul Sharma
Skills: Python, SQL, Excel, Machine Learning
Experience: 2 years in data analysis
Tools: Pandas, Power BI
"""
print(resume_text)
```

```python
def extract_skills(resume_text, skill_list):
    resume_text = resume_text.lower()
    extracted_skills = set()
    
    for skill in skill_list:
        if skill.lower() in resume_text:
            extracted_skills.add(skill.lower())
            
    return extracted_skills
```

```python
def calculate_match_percentage(job_skills, resume_skills):
    job_skills_set = set(skill.lower() for skill in job_skills)
    matched_skills = job_skills_set.intersection(resume_skills)
    
    match_percentage = (len(matched_skills) / len(job_skills_set)) * 100
    return match_percentage, matched_skills
```

```python
def shortlist_resume(match_percentage):
    if match_percentage >= 80:
        return "Shortlisted"
    else:
        return "Rejected"
```

```python
# Step 1: Extract skills from resume
resume_skills = extract_skills(resume_text, job_required_skills)

# Step 2: Calculate match %
match_percentage, matched_skills = calculate_match_percentage(
    job_required_skills,
    resume_skills
)

# Step 3: Shortlist decision
status = shortlist_resume(match_percentage)
```

```python
print("Job Title:", job_title)
print("Required Skills:", job_required_skills)
print("Resume Skills Found:", list(resume_skills))
print("Matched Skills:", list(matched_skills))
print("Match Percentage:", round(match_percentage, 2), "%")
print("Application Status:", status)
```
